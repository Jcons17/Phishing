***Phishing*** es un término informático que distingue a un conjunto de técnicas que persiguen el engaño a una víctima ganándose su confianza haciéndose pasar por una persona, empresa o servicio de confianza (suplantación de identidad de tercero de confianza), para manipularla y hacer que realice acciones que no debería realizar (por ejemplo revelar información confidencial o hacer click en un enlace).

---
### Técnicas de phishing

-***Phishing general, phishing tradicional, Bulk Phishing o Spray and pray.*** Consiste en la emisión masiva de correos electrónicos a usuarios. Estos correos suplantan a entidades de confianza (ejemplo bancos) y persiguen el engaño del usuario y la consecución de información. Por ejemplo en el mensaje se incluyen enlaces a dominios maliciosos. Para camuflar estos enlaces es habitual que el texto del enlace sea la URL correcta, pero el enlace en sí apunte al sitio malicioso.


-***Vishing.*** Es similar al phishing tradicional pero el engaño se produce a través de una llamada telefónica. El término deriva de la unión de dos palabras en inglés: ‘'voice'’ y ‘'phishing’'. Un ejemplo típico de uso de esta técnica es cuando un ciberdelincuente ha robado ya información confidencial a través de un ataque de phising, pero necesita la clave SMS o token digital para realizar y validar una operación. Es en ese momento el ciberdelincuente llama por teléfono al cliente identificándose como personal del banco y, con mensajes particularmente alarmistas, intenta de que el cliente revele el número de su clave SMS o token digital, que son los necesarios para autorizar la transacción.


-***Smishing.*** Es similar al phishing tradicional pero el engaño se produce a través mensajes de texto ya sean por SMS o mensajería instantánea (como WhatsApp). Un ejemplo típico de esta técnica es cuando el cliente recibe un mensaje de texto, donde el emisor se hace pasar por el banco, y le informan que se ha realizado una compra sospechosa con su tarjeta de crédito. A su vez, el texto solicita que se comunique con la banca por teléfono de la entidad financiera y le brinda un número falso. El cliente devuelve la llamada y es ahí cuando el ciberdelincuente, haciéndose pasar por el banco, solicita información confidencial para supuestamente cancelar la compra. En una variante de esta modalidad el mensaje también podría incluir un enlace a una ‘web’ fraudulenta para solicitar información sensible.12​

-***URL Phishing.*** Se trata de engañar al usuario haciendo que una URL de un sitio malicioso parezca la de un sitio confiable. A ellas a veces se accede de forma inadvertida al escribir nombres de dominio mal escritos que están muy cerca del dominio legítimo, o siguiendo un enlace malicioso que parece correcto, o por engaños al usar caracteres unicode parecidos difícilmente detectables, especialmente en dispositivos móviles, con pantallas más pequeñas y generalmente una resolución de pantalla inferior. Por ejemplo caracteres latinos con un punto bajo o la letra griega ómicron, "ο". . Otra forma de disfrazar enlaces es utilizar direcciones que contengan el carácter arroba: @, para posteriormente preguntar el nombre de usuario y contraseña (contrario a los). Por ejemplo, el enlace $http://www.google.com@members.tripod.com/$ puede engañar a un observador casual y hacerlo creer que el enlace va a abrir en la página de $www.google.com$, cuando realmente el enlace envía al navegador a la página de members.tripod.com (y al intentar entrar con el nombre de usuario de www.google.com, si no existe tal usuario, la página abrirá normalmente). Este método ha sido erradicado desde entonces en los navegadores de Mozilla e Internet Explorer.

En este proyecto trataremos de combatir el ***URL Phishing***, a través de una red neuronal. En Kaggle, existe una base de datos que presenta una cantidad considerable de URL's, que esta descompuesta en varios factores. Descargamos esta base de datos, y la abrimos con la libreria pandas

```python
import pandas as pd 
df = pd.read_csv("archive/phishing.csv")
df
```

---

Haremos un proceso de limpieza de la base de datos. Eliminaremos una fila o una columna que tenga mas del 30% de valores inválidos.

```python
por_filas = int(len(df)*0.3)
por_columnas = int(len(df.columns)*0.3)

print("Numero de entradas invalidas para eliminar filas: ",por_filas)
print("Numero de entradas invalidas para eliminar columnas: ",por_columnas)
```

Aplicamos un algoritmo para encontrar todos los valores inválidos por filas, sumar cuando un valor es invalido y cuando ese numero de entradas invalidas supere al 30% del total, se escribe el index de esa fila para posteriormente ser eliminado. Para las columnas, sera un proceso muy similar.

```python
import math
del_filas=[]
num_columns=len(df.columns)
for i in range(len(df)):
    m=0
    for j in range(num_columns):
        try:
            boolean = math.isnan(df.iloc[i,j])
            if boolean:
                m+=1  
        except:
            pass
    if m > por_filas:    
        del_filas.append(i)
```

```python
del_columnas_nan=[]
num_columns=len(df.columns)
for i in range(num_columns):
    m=0
    for j in range(len(df)):
        try:
            boolean = math.isnan(df.iloc[j,i])
            if boolean:
                m+=1  
        except:
            pass
    if m > por_columnas:    
        del_columnas_nan.append(i)
```

Imprimimos que columnas y filas serán eliminados.

```python
print(del_columnas_nan)
print(del_filas)
```

Eliminamos las columnas y filas necesarias. 

```python
del_columnas_nan = df.columns[del_columnas_nan].tolist()
```

```python
df = df.drop(del_filas,axis=0).reset_index(drop=True)
df = df.drop(del_columnas_nan, axis=1)
```

---

También para no sobrecargar las red neuronal, eliminaremos las columnas que tengan un indice de correlación mayor 0.9. Si dos columnas tienen una correlación alta, una de las dos son redundantes, por lo que podemos eliminar cualquiera de las dos. 

Cargamos la matriz de correlación entre los datos.

```python
corr_matrix = df.corr()
corr_matrix 
```

Recorremos cada una de las columnas y filas de la matriz de correlación, encontramos las que tenga una correlación absoluta arriba de 0.9, guardamos la correlación y las dos columnas que correlacionadas.

```python
del_columns=[]
for i in range(len(corr_matrix)):
    columns=[]
    for j in range(i):
        if i!=j:
            corr1=corr_matrix.iloc[i,j]
            if abs(corr1) > 0.9:
                del_columns.append((corr1,corr_matrix.columns[i],corr_matrix.columns[j]))
   # if len(columns) > 150:
    #    del_columns.append(df.columns[i])
print(del_columns)

```

```python
delete_columns = set(del_columns[i][2] for i in range(len(del_columns))) 
len(delete_columns)
print("las columnas a eliminar seran: \n",   str(delete_columns))
```

Eliminaremos las columnas que su type sea object:

```python
columns_object=[]
for i in range(len(df.columns)):
    if str(df.dtypes[i]) == "object":
        columns_object.append(df.columns[i])
columns_object
```

Eliminamos las columnas correlacionadas y las columnas objetos

```python
df = df.drop(delete_columns, axis=1)
df = df.drop(columns_object, axis=1)
```

Graficamos las correlaciones entre la nueva base de datos.

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig, ax = plt.subplots(figsize=(40,40))
sns.heatmap(df.corr(),annot=True,cmap='bwr',linewidths=0,annot_kws={"fontsize":6})
plt.xticks(size=10)
plt.yticks(size=10)
plt.savefig("corr.png")
plt.show()
```

Remapeamos la clase que nos dice si el registro efectivamente es phishing o no. Por defecto, nos dice que si el registro, en la columna *class* marca 1, es phishing, si marca -1 es phishing. Para eso usamos una función que transforma el -1 a 0 y el -1 lo deje a 0. Usaremos la función definida como:

$$
f(x) = \dfrac{x+1}{2}
$$

$$
f(1) = \dfrac{1+1}{2} = 1 
$$

$$
f(-1) = \dfrac{-1+1}{2}= 0  
$$

Esta función se la aplicamos a la columna class.

```python
df["class"]= df["class"].map(lambda x: int((x+1)/2))
```

Empezamos con los algoritmos deep learning.

- Importamos la librería Numpy para manejar arreglo de datos. 

- Importamos desde la librería sklearn, para dividir los datos de entrenamiento y prueba.



```python
import numpy as np
from sklearn.model_selection import train_test_split
```

Definimos la variable independiente como X_df y lo definimos como todas las columnas de la base de datos exceptuando por **class**. La variable dependiente sera Y_df que sera la columna **class**.

```python
X_df = df
X_df = X_df.drop(df.columns[-1], axis=1)
y_df = df[df.columns[-1]]
```

Dividimos los datos como training y test con un porcentaje del 25%

```python
x_tr, x_ts, y_tr, y_ts = train_test_split(X_df, y_df, test_size = 0.25, random_state = 42)
```

Establecemos los formatos de y_tr como array de numpy y establecemos su forma.

```python
y_tr = np.asarray(y_tr, dtype = np.int)
y_ts = np.asarray(y_ts, dtype = np.int)

y_tr = np.reshape(y_tr, [len(y_tr),1])
y_ts = np.reshape(y_ts, [len(y_ts),1])

print(x_ts.shape)
print(x_tr.shape)
print(y_ts.shape)
print(y_tr.shape)
```

Importamos *MinMAxScaler* que nos permitirá normalizar los datos de entrada.

```python
from sklearn.preprocessing import MinMaxScaler, Normalizer
# Training Data
x_tr = MinMaxScaler().fit_transform(x_tr)
print("Training Data :", x_tr.shape)

# Testing Data
x_ts = MinMaxScaler().fit_transform(x_ts)
print("Testing Data :", x_ts.shape)
```

Importamos todas las librerías necesarias para ejecutar el algoritmo que nos permitirá usar la red neuronal. 

```python
import tensorflow as tf
from matplotlib import pyplot as plt
from keras.models import Sequential
from keras.layers.core import Dense
from keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import TensorBoard
from tensorflow.keras.callbacks import ReduceLROnPlateau
```

Establecemos los hiperparametros. Después de algunas pruebas, concluí que la mejor opción sera usar 10000 iteraciones con ratio de aprendizaje de 0.01.

```python
# Hyperparameters
training_epochs = 200*5 # Total number of training epochs
learning_rate = 0.01 # The learning rate
momentum = 0.1
```

Creamos la arquitectura de nuestra red neuronal. 

- Capa de entrada con 31 inputs con la función de activación $tanh(x)= \dfrac{e^x - e^-x}{e^x + e^-x}$

- Capa oculta con 8 inputs con la función de activacion sigmoid.

- Capa de salida con 1 output de con la función de activación sigmoid.

Esta configuración no fue aleatoria, después de muchas pruebas exhaustivas (1era capa con softmax, segunda capa con 16 entradas, solo dos capas, etc) concluimos que está saca el máximo rendimiento de la red.  

```python
# create a model
def create_model():
    model = tf.keras.Sequential()
    # Hidden layer
    model.add(tf.keras.layers.Dense(len(X_df.columns), input_dim=len(X_df.columns),activation='tanh'))
    model.add(tf.keras.layers.Dense(8, input_dim=8,activation='sigmoid'))
    model.add(Dropout(0.5))
    # Output layer
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

    # Compile a model
    model.compile(loss='binary_crossentropy', 
                  optimizer=tf.keras.optimizers.SGD(learning_rate, momentum),
                  metrics=['accuracy'])
    return model
```

Creamos el objeto model y lo ejecutamos

```python
model = create_model()
model.summary()
```

Configuramos los callbacks de nuestra red neuronal. 

- EarlyStopping nos permite detener las iteraciones de las redes cuando ya no mejore.
- TensorBoard nos permite monitorear las iteraciones y visusalizar de mejor manera la red.
- ReduceLROnPlateau reduce el learning rate cuando deje de mejorar la red.

```python
my_callbacks = [
    EarlyStopping(monitor='loss',patience=10),
    TensorBoard(log_dir="./",
                write_graph=True,update_freq="epoch"),
    ReduceLROnPlateau(monitor='loss', factor=0.2,
                              patience=50, min_lr=0.001)
]
```

Aquí ejecutamos la red neuronal. Los datos a entrenar sera x_tr, y_tr. Los datos de validación seran x_ts, y_ts.

```python
results = model.fit(
    x_tr, y_tr,
    epochs= training_epochs,
    batch_size=32,
    validation_data = (x_ts, y_ts),
    callbacks=my_callbacks,
    verbose = 1
)
```

Imprimimos los resultados acerca del training set y del testing set.

```python
print("Evaluating on training set...")
(loss, accuracy) = model.evaluate(x_tr, y_tr, verbose=0)
print("loss={:.4f}, accuracy: {:.4f}%".format(loss,accuracy * 100))

print("Evaluating on testing set...")
(loss, accuracy) = model.evaluate(x_ts, y_ts, verbose=0)
print("loss={:.4f}, accuracy: {:.4f}%".format(loss,accuracy * 100))
```

Gráficamos la precisión y loss en contra de las iteraciones de la red.

```python
# summarize history for accuracy
fig = plt.subplots(figsize=(12,6))
plt.plot(results.history['accuracy'],color="b")
plt.plot(results.history['val_accuracy'],color="r")
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
# summarize history for loss
fig = plt.subplots(figsize=(12,6))

plt.plot(results.history['loss'])
plt.plot(results.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()


max_loss = np.max(results.history['loss'])
min_loss = np.min(results.history['loss'])
print("Maximum Loss : {:.4f}".format(max_loss))
print("")
print("Minimum Loss : {:.4f}".format(min_loss))
print("")
print("Loss difference : {:.4f}".format((max_loss - min_loss)))
```
